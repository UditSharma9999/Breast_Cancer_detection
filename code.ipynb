```python
M = pd.get_dummies(data["diagnosis"],drop_first=True)
M
```

## Preparing data for tranning

```python
data.drop(["diagnosis","Unnamed: 32"],axis=1,inplace=True)
data = pd.concat([data,M],axis=1)
data.head()
```

```python
X = data.drop("M",axis=1)
y = data["M"]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
scaler_features = scaler.transform(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(scaler_features, y, test_size = 0.2, random_state = 42)
```

## LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
Logistic_classifier = LogisticRegression(random_state = 0)
Logistic_classifier.fit(X_train,Y_train)
```

```python
Logistic_classifier_pred  = Logistic_classifier.predict(X_test)
Logistic_classifier_pred
```

```python
    print(classification_report(Y_test,Logistic_classifier_pred))
    print(accuracy_score(Y_test,Logistic_classifier_pred))
    Logistic_classifier_accuracy = accuracy_score(Y_test,Logistic_classifier_pred)*100
```

## KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
KNN_classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
KNN_classifier.fit(X_train, Y_train)
```

```python
KNN_classifier_pred = KNN_classifier.predict(X_test)
```

```python
print(classification_report(Y_test,KNN_classifier_pred))
print(accuracy_score(Y_test,KNN_classifier_pred))
KNN_classifier_accuracy = accuracy_score(Y_test,KNN_classifier_pred)*100
```

## SVC

```python
from sklearn.svm import SVC
SVC_classifier = SVC(kernel = 'linear', random_state = 0)
SVC_classifier.fit(X_train, Y_train)
```

```python
SVC_classifier_pred = SVC_classifier.predict(X_test)
```

```python
print(classification_report(Y_test,SVC_classifier_pred))
print(accuracy_score(Y_test,SVC_classifier_pred))
SVC_classifier_accuarcy = accuracy_score(Y_test,SVC_classifier_pred)*100
```

## GaussianNB

```python
from sklearn.naive_bayes import GaussianNB
GNB_classifier = GaussianNB()
GNB_classifier.fit(X_train, Y_train)
```

```python
GNB_classifier_pred = GNB_classifier.predict(X_test)
```

```python
print(classification_report(Y_test,GNB_classifier_pred))
print(accuracy_score(Y_test,GNB_classifier_pred))
GNB_classifier_accuracy = accuracy_score(Y_test,GNB_classifier_pred)*100
```

## DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
Treeclassifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
Treeclassifier.fit(X_train, Y_train)
```

```python
Treeclassifier_pred = Treeclassifier.predict(X_test)
```

```python
print(classification_report(Y_test,Treeclassifier_pred))
print(accuracy_score(Y_test,Treeclassifier_pred))
Treeclassifier_accuracy = accuracy_score(Y_test,Treeclassifier_pred)*100
```

## RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
RF_classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
RF_classifier.fit(X_train, Y_train)
```

```python
RF_classifier_pred = RF_classifier.predict(X_test)
```

```python
print(classification_report(Y_test,RF_classifier_pred))
print(accuracy_score(Y_test,RF_classifier_pred))
RF_classifier_accuracy = accuracy_score(Y_test,RF_classifier_pred)*100
```

## Accuracy Table

```python
from prettytable import PrettyTable
Table = PrettyTable(["Algorithm", "Accuracy"])
Table.add_row(["LogisticRegression",Logistic_classifier_accuracy])
Table.add_row(["KNeighborsClassifier", KNN_classifier_accuracy])
Table.add_row(["SVC", SVC_classifier_accuarcy])
Table.add_row(["GaussianNB",GNB_classifier_accuracy])
Table.add_row(["DecisionTreeClassifier", Treeclassifier_accuracy])
Table.add_row(["RandomForestClassifier", RF_classifier_accuracy])
print(Table)
```
